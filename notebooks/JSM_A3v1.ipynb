```python
# 4. Select the following columns- num_comments, num_shares_, num_reactions. [2]

fox_cols = fox[['num_comments','num_shares','num_reactions']]
```

```python
# 6. How many posts are there where status_type == 'link'? [1]

fox_stat = fox[fox['status_type'] == 'link']
print(fox_stat.shape)
print('There are 2666.')
```

```python
# 7. How many posts are there where status_type == 'video'? and number of shares is more than 500 [2]
fox_vids = fox[(fox['status_type'] == 'video') & (fox['num_shares'] > 500)]
print(fox_vids.shape)
print('There are 1869.')
```

```python
# 8. How many posts are there where number of comments is at least 2000? [1]

fox_coms = fox[fox['num_comments'] >= 2000]
print(fox_coms.shape)
print('There are 2639.')
```

```python
# 9. What is the maximum, average and minimum num_reactions,num_comments,num_shares,num_likes,num_loves,num_wows,num_hahas,
# num_sads,num_angrys per post? [3]
reacts = ['num_reactions','num_comments','num_shares','num_likes','num_loves','num_wows','num_hahas','num_sads','num_angrys']

fox[reacts].agg(['max','mean','min'])
```

```python
# 10. How many posts are there of different status_types? [Hint: Use groupby and count] [2]

fox_types = fox.groupby(by = 'status_type')['status_id'].count()
fox_types
print('There are 2666 links, 1119 photos, 107 statuses and 2096 videos.')
```

```python
# 11. What is the average num_reactions,num_comments,num_shares,num_likes,num_loves,num_wows,num_hahas,
# num_sads,num_angrys per post for each status_type? [Hint: groupby and mean] [2]

fox_reacts_by_type = fox.groupby(by = 'status_type')[reacts].mean()
fox_reacts_by_type
```

```python
# 15. Find the link type posts where the number of angry is higher than the average number of angry for a link type post.
# Store the result in a dataframe named 'angry_posts'. [3]

avg_angrys = fox[fox['status_type'] == 'link'].mean()
angry_posts = fox[(fox['status_type'] == 'link') & (fox['num_angrys'] > avg_angrys['num_angrys'])]

angry_posts.shape
```

```python
# 16. Find the average number of loves and average number of likes for the 'angry_posts'. [2]

angry_mean = angry_posts.agg('mean')
angry_mean
```

```python
# 17. Find the posts from 'angry_posts' where the num_likes is less than the average likes and num_loves is less than the
# average loves that you found in Question 16. [2]

less_angry_posts = angry_posts[(angry_posts['num_likes'] < angry_mean['num_likes']) & (angry_posts['num_loves'] < angry_mean['num_loves'])]
less_angry_posts
```
